```python
import pandas as pd
data = pd.read_excel(r'C:\Users\cxy\OneDrive\桌面\回国\实习\7-30\7-30 院长计划.xlsx',sheet_name='1-200')
#data = pd.read_excel(r'C:\Users\cxy\OneDrive\桌面\回国\实习\7-29\小规模测试.xlsx')
df = data.reset_index().drop(columns='index')
df['医院名称'] = df['医院名称'].astype('string')
df
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from fuzzywuzzy import fuzz
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

def setup_driver():
    path_to_chromedriver = r"D:\chromedriver-win64\chromedriver-win64\chromedriver.exe"
    service = Service(executable_path=path_to_chromedriver)
    options = webdriver.ChromeOptions()
    options.headless = True
    driver = webdriver.Chrome(service=service, options=options)
    return driver
first_time=True
def fetch_hospital_info(query_name, driver):
    global first_time
    try:
        url = f'https://aiqicha.baidu.com/s?q={query_name}'
        driver.get(url)
        wait = WebDriverWait(driver, 3)
        if first_time:
            time.sleep(10)  # 首次访问额外等待,记得验证，验证，验证！！！
            first_time = False  # 更新标记为非首次
        time.sleep(5)  # 强制等待，以确保所有内容都加载完成
        results = driver.find_elements(By.CSS_SELECTOR, '.title a')
        top_results = results[:3]  # 取前三个结果
        best_match = None
        highest_score = 0

        for result in top_results:
            hospital_name = result.text
            score = fuzz.ratio(query_name, hospital_name)
            if score > highest_score:
                highest_score = score
                best_match = result

        if best_match and highest_score > 60:
            hospital_name = best_match.text
            tags = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'tags-list')))
            if '事业单位' in tags.text:
                props = driver.find_element(By.CLASS_NAME, 'props')
                name_type = props.find_element(By.CSS_SELECTOR, 'span.pre-title').text
                leader_name = props.find_element(By.CSS_SELECTOR, 'span.legal-txt').text
                return query_name, hospital_name, name_type, leader_name,'公立'
            else:
                return query_name, hospital_name, name_type, leader_name,'非公立'
        else:
            return query_name, '无近似医院', '无', '无','无'
    except Exception as e:
        return query_name, '发生错误', '无', '无','无’


driver = setup_driver()
results = [fetch_hospital_info(name, driver) for name in df['医院名称']]
driver.quit()
    
results_df = pd.DataFrame(results, columns=['查询名称', '搜索到的医院名称', '领导类型','姓名','医院属性'])
results_df
```
